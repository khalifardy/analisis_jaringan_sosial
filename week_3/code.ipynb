```python
#read_csv file 
df = pd.read_csv('tweets-data/naturalisasi.csv')
df.head()
```

```python
#get user name 
username = list(set(df['username'].tolist()))
username_kumpulan = username[:15]
username_kumpulan
```

```python
count = 0 
url = 'https://api.x.com/2/users/by/username/'
bearer_token = "AAAAAAAAAAAAAAAAAAAAAAXSwAEAAAAAAy1vX36lEBtmOJ7HzyNXU9wJcY8%3DMZ5qe8DH8xWua8WIy2oHSjrXUpeDsMaRVkfetiHOIQvn5m7M0X"

path_params = 'BentVuijk'
params = {
    'user.fields': 'created_at,description,entities,id,location,name,pinned_tweet_id,profile_image_url,protected,public_metrics,url,username,verified,withheld'
}

headers = {
    "Authorization": f"Bearer {bearer_token}"
}

data_user_profile = {
    'id':[],
    'username':[],
    'name':[],
    'description':[],
    'location':[],
    'verified':[],
    'created_at':[],
    'profile_image_url':[],
    'followers_count':[],
    'following_count':[],
    'tweet_count':[],
    'like_count':[],
    'listed_count':[],
}
for user in username_kumpulan:
    response = requests.get(f'{url}{user}', headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        data_user_profile['id'].append(data['data']['id'])
        data_user_profile['username'].append(data['data']['username'])
        data_user_profile['name'].append(data['data']['name'])
        data_user_profile['description'].append(data['data']['description'])
        try:
            data_user_profile['location'].append(data['data']['location'])
        except:
            data_user_profile['location'].append(None)
        data_user_profile['verified'].append(data['data']['verified'])
        data_user_profile['created_at'].append(data['data']['created_at'])
        data_user_profile['profile_image_url'].append(data['data']['profile_image_url'])
        data_user_profile['followers_count'].append(data['data']['public_metrics']['followers_count'])
        data_user_profile['following_count'].append(data['data']['public_metrics']['following_count'])
        data_user_profile['tweet_count'].append(data['data']['public_metrics']['tweet_count'])
        data_user_profile['like_count'].append(data['data']['public_metrics']['like_count'])
        data_user_profile['listed_count'].append(data['data']['public_metrics']['listed_count'])
    else:
        print(response.status_code)
    count += 1
    time.sleep(30)
    if count % 3 == 0:
        time.sleep(60*15)

data_user_profile
```
